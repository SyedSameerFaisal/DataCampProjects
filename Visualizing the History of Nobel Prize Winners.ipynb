## What is the most commonly awarded gender and birth country?
- Store your answers as string variables top_gender and top_country.

```python
gender_distribution = npw['sex'].value_counts()
gender_distribution
```

```python
top_gender = gender_distribution.idxmax()
```

```python
print("The most commonly awarded gender is", top_gender)
```

```python
country_distribution = npw['birth_country'].value_counts()
country_distribution
```

```python
top_country = country_distribution.idxmax()
```

```python
print("The most commonly awarded country is", top_country)
```

## Which decade had the highest ratio of US-born Nobel Prize winners to total winners in all categories?
- Store this as an integer called max_decade_usa.

```python
npw['usa_winners'] = npw['birth_country'] == 'United States of America'
```

```python
npw['decade'] = (np.floor(npw['year'] / 10) * 10).astype(int)
```

```python
prop_usa_winners = npw.groupby('decade', as_index=False)['usa_winners'].agg(np.mean)
prop_usa_winners
```

```python
max_decade_usa = prop_usa_winners[prop_usa_winners['usa_winners'] == prop_usa_winners['usa_winners'].max()]['decade'].values[0]
```

```python
print("The decade that had the highest ratio of US-born Nobel Prize winners to total winners in all categories was the decade", max_decade_usa)
```

## Which decade and Nobel Prize category combination had the highest proportion of female laureates?
- Store this as a dictionary called max_female_dict where the decade is the key and the category is the value. There should only be one key:value pair.

```python
npw['female_winners'] = npw['sex'] == 'Female'
npw['female_winners']
```

```python
female_prop = npw.groupby(['decade', 'category'], as_index=False)['female_winners'].agg(np.mean)
```

```python
female_prop
```

```python
max_female_prop = female_prop.sort_values('female_winners', ascending=False).head(1)
```

```python
max_female_dict = {max_female_prop['decade'].values[0]: max_female_prop['category'].values[0]}
```

```python
print("The decade and Nobel Prize category combination that had the highest proportion of female laureates is", max_female_dict)
```

## Who was the first woman to receive a Nobel Prize, and in what category?
- Save your string answers as first_woman_name and first_woman_category.

```python
npw = npw.sort_values('year')
```

```python
first_woman_winner = npw[npw['sex'] == 'Female'].head(1)
```

```python
first_woman_name = first_woman_winner['full_name'].values[0]
first_woman_category = first_woman_winner['category'].values[0]
```

```python
print("The first woman to receive a Nobel Prize was", first_woman_name, "and her category was", first_woman_category)
```

```python
count = npw['full_name'].value_counts()
```

```python
repeat_list = count[count >= 2]
```

```python
repeat_list = list(repeat_list.index)
```
